# PROD

cleaning av prod datasett. fjerning av unødvendige kolonner

```python
df_prod.head()
```

```python
df_prod_red = df_prod.drop(columns=["Duplikat","Optimal_syklustid","Optm_antall","Kap_kategori","Sortering","VBA","Rangering","Vekt","Skifttimer_produkt","Skifttimer_skift","Maskintimer","Syklusantall","Avlesning","Syklustid","Driftstimer"])
```

```python
df_prod_red.head()
```

```python
df_prod_K_mach = df_prod_red[df_prod_red["Maskin"].str.contains("K")]
df_prod_F_mach = df_prod_red[df_prod_red["Maskin"].str.contains("F")]
df_prod_E_mach = df_prod_red[df_prod_red["Maskin"].str.contains("E")]
df_prod_D_mach = df_prod_red[df_prod_red["Maskin"].str.contains("D")]
```

### K maskiner

```python
df_prod_K_mach["Dato"] = pd.to_datetime(df_prod_K_mach["Dato"])
df_prod_K_mach.set_index("Dato",inplace=True)
```

```python
df_prod_K_mach.groupby(["Dato","Maskin"])["Prod_antall"].sum()
```

```python
## replace native values with 0
df_prod_K_mach[df_prod_K_mach["Prod_antall"]< 0] = 0

## remove extreme values
```

```python
df_prod_K_mach_seri = df_prod_K_mach.groupby([pd.Grouper(freq="M"),"Maskin"])["Prod_antall"].sum()
df_prod_K_mach_machine = df_prod_K_mach_seri.reset_index()
df_prod_K_mach_machine.set_index("Dato",inplace=True)
```

```python
fig = px.line(df_prod_K_mach_machine,y="Prod_antall",color="Maskin")
fig.show()
```

```python
df_prod_K_mach_machine.groupby([pd.Grouper(freq="M")])["Prod_antall"].sum().plot()
```

```python
#Save as Excel file

df_prod_K_mach_machine.to_excel("K_machines_produtcion.xlsx")
```

### F maskiner

```python
df_prod_F_mach["Dato"] = pd.to_datetime(df_prod_F_mach["Dato"])
df_prod_F_mach.set_index("Dato",inplace=True)
```

```python
df_prod_F_mach.groupby(pd.Grouper(freq="M"))["Prod_antall"].sum().plot()
```

```python
#Make data for each machine
df_prod_F_mach_seri = df_prod_F_mach.groupby([pd.Grouper(freq="M"),"Maskin"])["Prod_antall"].sum()
df_prod_F_mach_prod = df_prod_F_mach_seri.reset_index()
df_prod_F_mach_prod.set_index("Dato",inplace=True)
```

```python
fig = px.line(df_prod_F_mach_prod,y="Prod_antall",color="Maskin")
fig.show()
```

```python
#remove exreme values
df_prod_F_mach_prod_clean = df_prod_F_mach_prod

median = df_prod_F_mach_prod_clean["Prod_antall"].median()
std = df_prod_F_mach_prod_clean["Prod_antall"].std()
outliers = (df_prod_F_mach_prod_clean["Prod_antall"]-median).abs() > std
df_prod_F_mach_prod_clean[outliers] = np.nan
df_prod_F_mach_prod_clean.fillna(median, inplace=True)




```

```python
fig = px.line(df_prod_F_mach_prod_clean,y="Prod_antall",color="Maskin")
fig.show()
```

```python
fig = px.scatter(df_prod_F_mach_prod_clean,y="Prod_antall",color="Maskin",trendline="ols")
fig.show()
```

```python
df_prod_F_mach_prod_clean.to_excel("F_machines_produtcion.xlsx")
```
